#문항 sum을 예측하도록..

```python
#get factor
def get_factors(subset_df):
    factors = {}
    for i in subset_df.columns:
        if i[:3] not in factors:
            factors[i[:3]] = []         # key가 없으면 빈 리스트 생성
        factors[i[:3]].append(i)
    return factors
```

```python
def get_sum(subset_df):
    lis = []
    corr_df  = pd.DataFrame()
    for key, value in get_factors(subset_df).items():

        corr_df[key] = subset_df[value].sum(axis = 1)
    return corr_df
```

```python
#sum_df 설정...
origin_sum_df = get_sum(origin_df)
ga_sum_df = get_sum(Ga_df)
transformers_sum_df = get_sum(transformers_df)
CTT_sum_df = get_sum(CTT_df)
```

# Transformers short predict

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
origin_df[origin_df.isna().sum(axis = 1) != 0]
```

```python
X = transformers_df
y = origin_sum_df['EXT']
X_tensor = torch.tensor(X.values, dtype=torch.float32).to(device)
y_tensor = torch.tensor(y.values,dtype=torch.float32).to(device)
```

```python
# 데이터 분할
train_size = int(0.8 * len(X))
X_train, X_test = X_tensor[:train_size], X_tensor[train_size:]
y_train, y_test = y_tensor[:train_size], y_tensor[train_size:]
print("NaN in X_test:", torch.isnan(X_train).sum())
print("Inf in X_test:", torch.isinf(X_train).sum())
```

```python
# DataLoader
from torch.utils.data import TensorDataset, DataLoader

batch_size = 32
train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# 모델 정의
class RegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(25, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.net(x)

model = RegressionModel().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)

# 학습
epochs = 5
for epoch in range(epochs):
    model.train()
    for xb, yb in train_loader:
        preds = model(xb)
        loss = criterion(preds, yb.unsqueeze(-1))
        if torch.isnan(loss):
            print("❌ NaN 발생! xb 또는 yb 값 확인 필요")
            print("xb NaN:", torch.isnan(xb).sum())
            print("yb NaN:", torch.isnan(yb).sum())
            break
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(epoch)

# 평가
model.eval()
with torch.no_grad():
    y_pred = model(X_test).to('cpu').numpy()
    y_true = y_test.to('cpu').numpy()
# 1차원으로 강제 변환
y_true = y_true.flatten()
y_pred = y_pred.flatten()


print("NaN in y_pred:", np.isnan(y_pred).sum())
print("NaN in y_true:", np.isnan(y_true).sum())
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)
mape = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-8))) * 100

print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²: {r2:.4f}")
print(f"MAPE: {mape:.2f}%")
```
