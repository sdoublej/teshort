```python
import pandas as pd
# Mahalanobis 거리 (전체 문항 기준)
import numpy as np
from scipy.spatial import distance
```

# load df! , and i drop the zero

```python
df = pd.read_csv("/content/drive/MyDrive/학부연구생/C프로젝트/C프로젝트 최종정리/df.csv", index_col = 0)
```

```python
print(f'len(df): {len(df)}')
# zero 무응답만 빼버렸음
drop_index = df[(df == 0 ).sum(axis = 1)  != 0].index
print(f'len(drop_index): {len(drop_index)}')
droped_df = df.drop(drop_index)
dropna_index = droped_df[df.isna().sum(axis = 1) != 0].index
print(f'len(dropna_index): {len(dropna_index)}')
droped_df = droped_df.drop(dropna_index)
print(f'len(droped_df): {len(droped_df)}')
```

```python
reversed = pd.read_csv('/content/drive/MyDrive/학부연구생/C프로젝트/C프로젝트 최종정리/Revised_and_Original_Items.csv').Item.values
```

# reverse definition

```python
def reverser(x):
    return 6 - x
```

```python
reversed
```

#무조건 한 번만 실행!

```python
for i in reversed:
    droped_df[i] = droped_df[i].map(reverser)
```
