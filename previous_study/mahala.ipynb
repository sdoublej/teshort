```r
# 예제 데이터 (예: 50문항짜리 심리 설문지 응답 데이터프레임)
# 실제로는 df <- read.csv("your_data.csv") 등으로 불러오세요.
set.seed(123)
df <- read.csv("/content/sample_data/first_droped_reverse_df.csv", row.names = 1)  # 100명, 50문항

# Mahalanobis 거리 계산
mahal_dists <- mahalanobis(df, colMeans(df), cov(df))

# p-value 계산 (카이제곱 분포 기준)
p_vals <- pchisq(mahal_dists, df = ncol(df), lower.tail = FALSE)

# 이상치 플래그 지정 (예: p < .001이면 이상치로 간주)
df$mahal_dist <- mahal_dists
df$p_value <- p_vals
df$outlier_flag <- ifelse(p_vals < 0.001, "Outlier", "Normal")

# 결과 요약
table(df$outlier_flag)
```
